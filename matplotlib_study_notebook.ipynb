```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0, 10, 500)
y1= np.sin(x)
y2= np.sin(x) + 0.5

plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)
plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='red')
plt.title("Filled Area Between Curves")
plt.show()
```

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]


# cria grfico de lina, anos no eixo x, gdp no eixo y
plt.plot(years, gdp, color='green', marker='o', linestyle= 'solid')

# adiciona um ttulo
plt.title("GDP Nominal")

# adiciona um ttulo
plt.title("GDP Nominal")

# adiciona um selo no eixo y
plt.ylabel("Bilhes de $")
plt.show()
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]


# barras possuem o tamanho padro de 0.8, ento adicionaremos 0.1 s
# coordenadas  esquerda para que cada barra seja centralizada

xs = [i + 0.1 for i, _ in enumerate(movies)]

# barras de grfico com as coordenadas x  esquerda [xs], alturas[num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# de Premiaes")
plt.title("Meus Filmes Favoritos")

# nomeia o eixo x com nomes de filmes na barra central
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

plt.show()
```

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar(histogram.keys(), # move cada barra para a esquerda em 4
        histogram.values(),                 # d para cada barra sua altura correta
        8)                                  # d pra cada barra a largura de 8

plt.axis([-5, 105, 0, 5])                   # eixo x de -5 at 105, eixo de 0 at 5

plt.xticks([10 * i for i in range(11)]) # rtulos do eixo x em 0, 10, , 100
plt.xlabel("Decil")
plt.ylabel("# de Alunos")
plt.title("Distribuio das Notas do Teste 1")
plt.show()
```
