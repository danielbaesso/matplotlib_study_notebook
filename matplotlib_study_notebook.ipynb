```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0, 10, 500)
y1= np.sin(x)
y2= np.sin(x) + 0.5

plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)
plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='red')
plt.title("Filled Area Between Curves")
plt.show()
```

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]


# cria grfico de lina, anos no eixo x, gdp no eixo y
plt.plot(years, gdp, color='green', marker='o', linestyle= 'solid')

# adiciona um ttulo
plt.title("GDP Nominal")

# adiciona um ttulo
plt.title("GDP Nominal")

# adiciona um selo no eixo y
plt.ylabel("Bilhes de $")
plt.show()
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]


# barras possuem o tamanho padro de 0.8, ento adicionaremos 0.1 s
# coordenadas  esquerda para que cada barra seja centralizada

xs = [i + 0.1 for i, _ in enumerate(movies)]

# barras de grfico com as coordenadas x  esquerda [xs], alturas[num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# de Premiaes")
plt.title("Meus Filmes Favoritos")

# nomeia o eixo x com nomes de filmes na barra central
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

plt.show()
```

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar(histogram.keys(), # move cada barra para a esquerda em 4
        histogram.values(),                 # d para cada barra sua altura correta
        8)                                  # d pra cada barra a largura de 8

plt.axis([-5, 105, 0, 5])                   # eixo x de -5 at 105, eixo de 0 at 5

plt.xticks([10 * i for i in range(11)]) # rtulos do eixo x em 0, 10, , 100
plt.xlabel("Decil")
plt.ylabel("# de Alunos")
plt.title("Distribuio das Notas do Teste 1")
plt.show()
```

```python
# Exemplo de um grfico ruim, eixo y no comea no 0

mentions = [500,505]
years = [2013, 2014]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# de vezes que ouvimos algum dizer 'data science'")

# se voc no fizer isso, matplotlib nomear o eixo x de 0, 1

# e ento adiciona a +2.013e3 para fora do canto (matplotlib feio!)
plt.ticklabel_format(useOffset=False)

# enganar o eixo y mostra apenas a parte acima de 500

plt.axis([2012.5, 2014.5, 499, 506])
plt.title("Olhe o \"Grande\" Aumento!")
plt.show()

```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# podemos fazer mltiplas chamadas para plt.plot
# para mostrar mltiplas sries no mesmo grfico
plt.plot(xs, variance, 'g-', label='variance') # linha verde slida
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # linha de ponto tracejado vermelho
plt.plot(xs, total_error, 'b:', label='total error') # linha com pontilhado azul

# porque atribumos rtulos para cada srie
# podemos obter uma legenda gratuita
# loc=9 significa "top center"

plt.legend(loc=9)
plt.xlabel("complexidade do modelo")
plt.title("Compromisso entre Polarizao e Varincia")
plt.show()
```

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# nomeia cada posio
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count), # coloca o rtulo com sua posio
                 xytext=(5, -5), # mas compensa um pouco
                 textcoords = 'offset points')
    
plt.title("Minutos Dirios vs. Nmero de Amigos")

plt.xlabel("# de amigos")
plt.ylabel("minutos dirios passados no site")
plt.show()
```

# Organize

You can plot several data on the same figure, but you can also split a figure in several subplots (named Axes):

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()

# ax.set_title("Grfico de Linha")
# ax.set_xlabel("Eixo X")
# ax.set_ylabel("Eixo Y")

x = np.linspace(0, 10, 100)
y1, y2 = np.sin(x), np.cos(x)

ax.plot(x, y1,x, y2)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(x, y1, color="C1")
ax2.plot(x, y2, color="C0")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(y1, x, color="C1")
ax2.plot(y2, x, color="C0")
```
